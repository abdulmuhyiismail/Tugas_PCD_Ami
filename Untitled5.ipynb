```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan gambar
def display_images(images, titles, cmap=None):
    plt.figure(figsize=(15, 5))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Load gambar berwarna dan ubah menjadi grayscale
image_color = cv2.imread("/content/Rahman.jpeg")  # Ganti "image.jpg" dengan path gambar Anda
image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)  # Konversi ke RGB
image_gray = cv2.cvtColor(image_color, cv2.COLOR_RGB2GRAY)

# Filter Low-pass (Gaussian Blur)
low_pass_color = cv2.GaussianBlur(image_color, (15, 15), 0)
low_pass_gray = cv2.GaussianBlur(image_gray, (15, 15), 0)

# Filter High-pass (menggunakan kernel Laplacian)
laplacian_color = cv2.Laplacian(image_color, cv2.CV_64F)
laplacian_color = cv2.convertScaleAbs(laplacian_color)
laplacian_gray = cv2.Laplacian(image_gray, cv2.CV_64F)
laplacian_gray = cv2.convertScaleAbs(laplacian_gray)

# Filter High-boost (kombinasi gambar asli dan high-pass)
boost_factor = 1.5
high_boost_gray = cv2.addWeighted(image_gray, boost_factor, laplacian_gray, 1, 0)
high_boost_color = cv2.addWeighted(image_color, boost_factor, laplacian_color, 1, 0)

# Tampilkan hasil
display_images(
    [image_gray, low_pass_gray, laplacian_gray, high_boost_gray],
    ["Grayscale", "Low-pass (Blur)", "High-pass (Edges)", "High-boost"],
    cmap="gray",
)

display_images(
    [image_color, low_pass_color, laplacian_color, high_boost_color],
    ["Original Color", "Low-pass (Blur)", "High-pass (Edges)", "High-boost"],
)
```
